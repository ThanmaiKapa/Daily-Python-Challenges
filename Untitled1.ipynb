```python
class Solution(object):
    def maxArea(self, height):
        i=0
        j=len(height)-1
        most_water = 0
        while i<j:
            most_water = max(most_water, (j-i)*min(height[i],height[j]))
            if height[i]<height[j]:
                i+=1
            else:
                j-=1
        return most_water
h = list(map(int,input("Enter List: ").split()))
solution = Solution()
print(solution.maxArea(h))
```
